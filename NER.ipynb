```python
#CODE FOR THE EXTRACTION OF LAUNCH DATE FROM PDF FILES ORGANIZED MARKET AND UPDATING THE LAUNCH DATE IN THE DATABASE
import fitz  
import re
import glob
import os
from pymongo import MongoClient

client = MongoClient('mongodb+srv://moiz121:test123456@cluster0.lq8grwq.mongodb.net/')  
db = client['KDPWCCP']  
collection = db['RefineddataOrganizedMarket']  

directory_path = "D:\\PDF\\'OrganizedMarket"
pdf_files = glob.glob(os.path.join(directory_path, "*.pdf"))

def extract_resolution_number(file_name):
   
    match = re.match(r'(\d{4})-(\d{4})[-_]?en\.pdf', file_name)
    if match:
        resolution_num, year = match.groups()
        return f"Resolution No. {int(resolution_num)} / {year}"
    return None

for pdf_file in pdf_files:
    
    with fitz.open(pdf_file) as pdf:
        all_text = ""
        for page in pdf:
            all_text += page.get_text() or ""  

   
    match = re.search(r'(dated\s+)(\d{1,2}\s\w+\s\d{4})', all_text, re.IGNORECASE)

    if match:
        launch_date = match.group(2)
        file_name = os.path.basename(pdf_file)
        print(f"Launch date from {file_name}: {launch_date}")

        resolution_number = extract_resolution_number(file_name)
        print(f"Extracted resolution number: {resolution_number}")

        if resolution_number:
            print(f"Querying for resolution_number: {resolution_number}")

            
            result = collection.update_one(
                {"resolution_number": resolution_number}, 
                {"$set": {"launch_date": launch_date}},  
                upsert=False  
            )

            if result.matched_count > 0:
                print(f"Updated document for {file_name}")
            else:
                print(f"No matching document found for {file_name}")
        else:
            print(f"No resolution number found for {file_name}")

    else:
        print(f"No launch date found in {os.path.basename(pdf_file)}")
```

```python
import fitz
import glob
import os
from pymongo import MongoClient
from transformers import BartForConditionalGeneration, BartTokenizer
import re

model_name = "nickmuchi/fb-bart-large-finetuned-trade-the-event-finance-summarizer"
model = BartForConditionalGeneration.from_pretrained(model_name)
tokenizer = BartTokenizer.from_pretrained(model_name)

def connect_to_mongodb(database_name, collection_name):
    client = MongoClient("mongodb+srv://moiz121:test123456@cluster0.lq8grwq.mongodb.net/")
    db = client[database_name]
    collection = db[collection_name]
    return collection

def generate_summary(text):
    input_text = "summarize: " + text
    input_ids = tokenizer(input_text, return_tensors="pt", max_length=1024, truncation=True).input_ids
    output_ids = model.generate(
        input_ids,
        max_length=250,
        num_beams=4,
        length_penalty=1.0,
        early_stopping=True,
        min_length=50
    )
    summary = tokenizer.decode(output_ids[0], skip_special_tokens=True)
    if not summary.endswith('.'):
        summary += '.'
    return summary

def extract_organized_text_from_pdf(pdf_path):
    document = fitz.open(pdf_path)
    organized_text = ""
    
    for page_num in range(len(document)):
        page = document.load_page(page_num)
        blocks = page.get_text("dict")["blocks"]
        
        for block in blocks:
            if block.get('type') == 0:  
                block_text = ""
                for line in block.get("lines", []):
                    for span in line.get("spans", []):
                        block_text += span.get("text", "") + " "
                organized_text += block_text.strip() + "\n"
            elif block.get('type') == 1:  
                for line in block.get("lines", []):
                    row = ""
                    for span in line.get("spans", []):
                        row += span.get("text", "") + " | "
                    organized_text += row.strip() + "\n"
    
    document.close()
    organized_text = organized_text.replace('\n', ' ')
    organized_text = organized_text.strip()
    return organized_text

def extract_resolution_number_from_filename(file_name):
    match = re.match(r'(\d{4})-(\d{4})[-_]?en\.pdf', file_name)
    if match:
        resolution_num, year = match.groups()
        return f"Resolution No. {int(resolution_num)} / {year}"
    return None

def update_mongodb_with_pdf_data(refined_collection, pdf_directory):
    pdf_files = glob.glob(os.path.join(pdf_directory, "*.pdf"))
    print(f"Found {len(pdf_files)} PDF files.")

    for pdf_file in pdf_files:
        pdf_filename = os.path.basename(pdf_file)
        resolution_number = extract_resolution_number_from_filename(pdf_filename)
        
        if resolution_number:
            print(f"Processing file: {pdf_filename}")

            # Extracting organized text from PDF
            extracted_text = extract_organized_text_from_pdf(pdf_file)
            if extracted_text:
                print(f"Extracted text from {pdf_filename}.")

                summary = generate_summary(extracted_text)

                
                result = refined_collection.update_one(
                    {"resolution_number": resolution_number},
                    {
                        "$set": {
                            "Text_extracted_from_pdf": extracted_text,
                            "AI_summary_of_PDF": summary
                        }
                    }
                )

                if result.matched_count > 0:
                    print(f"Updated document for resolution_number: {resolution_number}")
                else:
                    print(f"No matching document found for resolution_number: {resolution_number}")
            else:
                print(f"No text extracted from {pdf_filename}.")
        else:
            print(f"Filename {pdf_filename} does not match the expected pattern.")

def main():
    database_name = "KDPWCCP"
    refined_collection_name = "RefineddataOrganizedMarket"
    pdf_directory = "D:\\PDF\\'OrganizedMarket"

    refined_collection = connect_to_mongodb(database_name, refined_collection_name)

    update_mongodb_with_pdf_data(refined_collection, pdf_directory)

if __name__ == "__main__":
    main()
```
